```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
%matplotlib inline
```

```python
# Generate training data

def true_fun(X):
    return np.cos(1.5 * np.pi * X)

np.random.seed(0)

n_samples = 15
degrees = [1, 4, 15]

X_train = np.sort(np.random.rand(n_samples))
Y_train = true_fun(X_train) + np.random.randn(n_samples) * 0.1
```

```python
# Plot training data and data generating function

x = np.linspace(0, 1, 100)
y = true_fun(x)

plt.ylim(-1.2, 1.2)
plt.scatter(X_train, Y_train)
plt.plot(x, y)
```

```python
# Fit 1st degree polynomial

degree = 1

model = Pipeline([("poly_feature", PolynomialFeatures(degree, include_bias=False)),
                  ("linear_regression", LinearRegression())])

model.fit(X_train.reshape(-1, 1), Y_train)

y = model.predict(x.reshape(-1, 1))

plt.ylim(-1.2, 1.2)
plt.scatter(X_train, Y_train)
plt.plot(x, y)
```

```python
# Fit 3d degree polynomial

degree = 3

model = Pipeline([("poly_feature", PolynomialFeatures(degree, include_bias=False)),
                  ("linear_regression", LinearRegression())])

model.fit(X_train.reshape(-1, 1), Y_train)

y = model.predict(x.reshape(-1, 1))

plt.ylim(-1.2, 1.2)
plt.scatter(X_train, Y_train)
plt.plot(x, y)
```

```python
# Fit 15th degree polynomial

degree = 15

model = Pipeline([("poly_feature", PolynomialFeatures(degree, include_bias=False)),
                  ("linear_regression", LinearRegression())])

model.fit(X_train.reshape(-1, 1), Y_train)

y = model.predict(x.reshape(-1, 1))

plt.ylim(-1.2, 1.2)
plt.scatter(X_train, Y_train)
plt.plot(x, y)
```
